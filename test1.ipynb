```python
%matplotlib inline

# print("Hello world!")
import numpy as np
import matplotlib.pyplot as plt
```

Building the matrix representing the Laplacian operator and computing eigenvalues & eigenvectors

Obviously this requires some preliminary work (basis functions & integration)

```python
N = 100
L = 1
A = np.zeros((N-1,N-1), float)
for i in range(N-1):
    A[i][i] = 2*N**2/L**2
for i in range(N-2):
    A[i][i+1] = -N**2/L**2
    A[i+1][i] = -N**2/L**2
eigenValues,eigenVectors = np.linalg.eig(A)
```

Now sorting by increasing eigenvalues

```python
# idx = eigenValues.argsort()[::-1] reverse order  
idx = eigenValues.argsort()   
eigenValues = eigenValues[idx]
eigenVectors = eigenVectors[:,idx]
```

filling the initial and final zero
and comparing with theoretical values

```python
z = np.zeros((1,N-1),float)
eigenVectors = np.concatenate((z,eigenVectors,z), axis=0)
eigenValuesTheoretical = [np.pi**2/L**2*x**2 for x in range(N)]
```

Finally one has to exploit the data

plot eigenvectors

```python
for i in range (3):
    #plt.plot(eigenVectors[:,N-i-2]) last eigenvalues
    plt.plot(eigenVectors[:,i])
plt.title('Eigenvectors')
plt.show()
```

plot eigenvalues

```python
plt.plot(eigenValues)
plt.plot(eigenValuesTheoretical)
plt.title('Eigenvalues')
plt.show()
```
